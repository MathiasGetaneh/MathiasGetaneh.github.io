```python
import folium
import pandas as pd
city_data = pd.DataFrame({
    'lat':[38.85, 38.88, 38.7, 38.71, 38.82, 38.89, 38.75, 39.11, 38.3, 38.92],
    'lon':[-77.3, -77.1, -77.22, -77.1, -77.09, -77.45, -77.48, -77.55, -77.49, -78.19],
    'name':['Fairfax', 'Arlington', 'Lorton', 'Mount Vernon', 'Alexandria', 'Chantilly', 'Manassas', 'Leesburg', 'Fredericksburg', 'Front Royal'],
    'value':[0,1,4,3,2,9,6,8,5,7]
}, dtype=str)
city_data
```

```python
source = (
    '&copy; <a href="https://www.openstreetmap.org/copyright">OpenStreetMap</a> '
    'contributors, &copy; <a href="https://cartodb.com/attributions">CartoDB</a>'
)
no_labels = "https://{s}.basemaps.cartocdn.com/light_nolabels/{z}/{x}/{y}.png"
nova_map = folium.Map(location = [38.7, -77.4], tiles = no_labels, attr = source, zoom_start = 9)
for i in range(0, len(city_data)):
    folium.CircleMarker(
        location=[city_data.iloc[i]['lat'], city_data.iloc[i]['lon']],
        radius = 1,
        color = 'black',
        weight = 3,
        popup=city_data.iloc[i]['name'],
    ).add_to(nova_map)
    name = city_data.iloc[i]['name']
    length = len(name)/2*5
    folium.Marker(
        location=[city_data.iloc[i]['lat'], city_data.iloc[i]['lon']],
        icon = folium.DivIcon(
            icon_size=(130, 36),
            icon_anchor=(length, 0),
            html = f'<div style="font-size: 8pt; color: black;">{name}</div>',
        )
    ).add_to(nova_map)
nova_map
```

```python
nova_map.save("cities.html")
```

```python
city_orders = []
for i in range(0,len(city_data)):
    first_lon = 0
    first_lat = 0
    first_name = ''
    second_lon = 0
    second_lat = 0
    second_name = ''
    for j in range(0, len(city_data)):
        if (city_data.iloc[j]['value'] == str(i)):
            first_lon = float(city_data.iloc[j]['lon'])
            first_lat = float(city_data.iloc[j]['lat'])
            first_name = city_data.iloc[j]['name']            
        if (city_data.iloc[j]['value'] == str(i+1)) or (city_data.iloc[j]['value'] == str(i+1-10)):
            second_lon = float(city_data.iloc[j]['lon'])
            second_lat = float(city_data.iloc[j]['lat'])
            second_name = city_data.iloc[j]['name']
    city_orders.append([first_lat,first_lon])
    if (i == len(city_data)-1):
        city_orders.append([second_lat, second_lon])

print(city_orders)
```

```python
folium.PolyLine(
    locations = city_orders,
    color = '#FF0000',
    weight = 3,
    tooltip = "TSP"
).add_to(nova_map)
nova_map
```
